```python
import numpy as np

print(summe_quadratische_abstaende_100)
punkte = np.array(list(zip(range(1, 101), summe_quadratische_abstaende_100)))
print(summe_quadratische_abstaende_100)
start_punkt = punkte[0]
end_punkt = punkte[-1]
linien_vektor = end_punkt - start_punkt
punkt_vektoren = punkte - start_punkt


projizierte_laenge = np.dot(punkt_vektoren, linien_vektor) / np.linalg.norm(linien_vektor)
projizierte_punkte = np.outer(projizierte_laenge, linien_vektor) / np.linalg.norm(linien_vektor)**2
senkrechte_distanzen = np.linalg.norm(punkt_vektoren - projizierte_punkte, axis=1)

optimale_cluster_anzahl = senkrechte_distanzen.argmax() + 1

print(f"Die optimale Anzahl von Clustern ist: {optimale_cluster_anzahl}")
```

```python
import math

def berechne_distanz(px, py, x1, y1, x2, y2):
    # Berechnung der Distanz eines Punktes (px, py) von einer Geraden, definiert durch die Punkte (x1, y1) und (x2, y2)
    zaehler = abs((y2 - y1) * px - (x2 - x1) * py + x2 * y1 - y2 * x1)
    nenner = math.sqrt((y2 - y1) ** 2 + (x2 - x1) ** 2)
    return zaehler / nenner

def finde_entferntesten_punkt(y_werte):
    # Sicherstellen, dass die Liste mindestens zwei Werte enthält
    if len(y_werte) < 2:
        return None

    # Anfangs- und Endpunkt der Geraden (X-Werte entsprechen den Indizes)
    x1, y1 = 1, y_werte[0]
    x2, y2 = len(y_werte) , y_werte[-1]

    max_distanz = 0
    entferntester_punkt = None

    for index, py in enumerate(y_werte):
        px = index +1  # X-Wert entspricht dem Index
        distanz = berechne_distanz(px, py, x1, y1, x2, y2)
        if distanz > max_distanz:
            max_distanz = distanz
            entferntester_punkt = (px, py)

    return entferntester_punkt, max_distanz



entferntester_punkt, distanz = finde_entferntesten_punkt(summe_quadratische_abstaende_100[:10])
print(f"Entferntester Punkt: {entferntester_punkt} mit einer Distanz von: {distanz}")
```

```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np

def finde_optimale_strategie(df_merged):
    # Konvertiere den DataFrame, falls notwendig, von einem String-Format in ein numerisches Format
    # und ersetze '...' durch spezifische Spaltennamen, falls vorhanden
    df_merged = df_merged.apply(pd.to_numeric, errors='coerce', axis=1)
    
    # Initialisiere ein Dictionary, um die beste Strategie für jeden Cluster zu speichern
    beste_strategien = {}

    # Iteriere über jeden einzigartigen Cluster im DataFrame
    for cluster in df_merged['Cluster'].unique():
        # Filtere den DataFrame für den aktuellen Cluster
        cluster_df = df_merged[df_merged['Cluster'] == cluster]
        
        # Entferne die Spalte "Cluster", da sie nicht für die Berechnungen benötigt wird
        cluster_df = cluster_df.drop(columns=['Cluster'])
        
        # Initialisiere eine Variable für die minimale durchschnittliche Zeit und den Strategienamen
        min_durchschnittszeit = np.inf
        beste_strategie = None
        
        # Iteriere über die Strategien (Spalten) im cluster_df
        for strategie in cluster_df.columns[:-1]:  # Letzte Spalte (Cluster) bereits entfernt
            # Berechne die durchschnittliche Zeit für die Strategie, ignoriere Zeiten >= 500
            gueltige_zeiten = cluster_df[strategie][cluster_df[strategie] < 500]
            
            # Prüfe, ob die Strategie alle Probleme im Cluster lösen kann
            if len(gueltige_zeiten) == len(cluster_df):
                durchschnittszeit = gueltige_zeiten.mean()
                
                # Aktualisiere die beste Strategie, falls diese Strategie schneller ist
                if durchschnittszeit < min_durchschnittszeit:
                    min_durchschnittszeit = durchschnittszeit
                    beste_strategie = strategie
                    
        # Speichere die beste Strategie für den Cluster
        beste_strategien[cluster] = beste_strategie
    
    return beste_strategien

# Anwenden der Funktion
optimale_strategien = finde_optimale_strategie(df_merged)
print(optimale_strategien)
```

```python
 
    
def finde_optimale_strategie(df, max_zeit=500):
    optimale_strategien = {}
    df = df.apply(pd.to_numeric, errors='coerce', axis=1)
    for cluster in df['Cluster'].unique():
        cluster_df = df[df['Cluster'] == cluster]
        strategien = df.columns[1:-2]  
        durchschnittszeiten = {}
        for strategie in strategien:
            zeiten = cluster_df[strategie]
            zeiten = zeiten[zeiten < max_zeit]  
            if len(zeiten) == len(cluster_df): 
                durchschnittszeiten[strategie] = zeiten.mean()
        if durchschnittszeiten:
            optimale_strategie = min(durchschnittszeiten, key=durchschnittszeiten.get)
            optimale_strategien[cluster] = optimale_strategie
        else:
            optimale_strategien[cluster] = None  
    
    return optimale_strategien


optimale_strategien = finde_optimale_strategie(df_merged)
print(optimale_strategien)
```
